# Round 1: Using Wine Dataset


##load dataset and preprocess it

```python
from sklearn.datasets import load_wine
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression


dataset = load_wine()
dataset.keys()
```

```python
dataset.data.shape
```

```python
dataset.data[0]
```

```python
dataset.target
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df.head()
```

```python
df.describe()
```

```python
X = df
y = dataset.target
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
X
```

## Train a simple logistic regression classifier

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

## Conclusion: With 100% information (13 features), we can achieve 97.2% accuracy

# Round 2: Use PCA with 0.95 for analysis

## Keep 95% information PCA for dimension deduction

```python
X
```

```python
X.shape
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```

## Conclusion, with 95% information (10 features), we achieved 97.2% accuracy (equally good as using 13 features).
