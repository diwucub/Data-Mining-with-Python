# Data Visualization

It is hard to digest many data at the same time. Data visualizaiton is a way to make the process simple and straightforward.

For example, the sorting algorithms we have learned so far, are hard to beginners. However, a data visualizaiton can help the comparison much simpler: https://www.youtube.com/watch?v=kPRA0W1kECg

Because Data Visualization is so important and essential in today's business, pandas as a data-driven package provides built-in plotting functions. We will learn some basic usage of them today.


## import packages

```
import numpy as np
import pandas as pd
```

```
df = pd.read_csv('/work/superstore.csv')
df.drop('Row ID', axis = 1, inplace = True)
df.head()
```

```
df.info()
```

## Scatter Plots

```
df.plot(x = 'Sales', y = 'Profit', kind = 'scatter')
```

## Line Plots

```
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Order Year'] = df['Order Date'].apply(lambda d: d.year)
```

```
df.plot(x = 'Order Year', y = 'Sales', kind = 'line')
```

```
df2 = df.groupby('Order Year').sum()
df2
```

```
df2.plot(y = 'Sales', kind = 'line')
```

## Area Plots

```
df2.plot(y = 'Sales', kind = 'area')
```

```
df['Order Month'] = df['Order Date'].apply(lambda d: d.month)
df3 = df.groupby(['Order Year','Order Month']).sum()
df3
```

```
df3.plot(y = 'Sales', kind = 'area')
```

## Bar Charts

```
df3.plot(y = 'Sales', kind = 'bar')
```

```
df2.plot(y = 'Sales', kind = 'bar')
```

```
df3.plot(y = 'Sales', kind = 'barh')
```

```
df2.plot(y = 'Sales', kind = 'barh')
```

## Histograms

```
df['Sales'].plot(kind = 'hist')
```

```
df['Sales'].plot(kind = 'hist', bins = 100)
```

```
df['Quantity'].plot(kind = 'hist')
```

```
df['Quantity'].plot(kind = 'kde')
```

## Box plot

```
df['Sales'].plot(kind = 'box')
```

```
df['Quantity'].plot(kind = 'box')
```

## Pie Charts

```
df.groupby('Ship Mode')['Ship Mode'].count().plot(kind = 'pie')
```

## Color map

```
df.plot.scatter(x = 'Order Date', y = 'Sales', c = 'Profit')
```

```
df.plot.scatter(x = 'Sales', y = 'Profit', c = 'Discount')
```
